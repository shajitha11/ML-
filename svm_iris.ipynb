```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target']=iris.target
df.head(2)
```

```python
df['target names']=df.target.apply(lambda x : iris.target_names[x])
df.head(2)
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="purple",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="yellow",marker='+')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='*')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='*')
```

```python
X=df.drop(['target','target names'],axis='columns')
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

parameters tuning for svc(regularization(c),gamma,kernel)

```python
model_c=SVC(C=1)
model_c.fit(X_train,y_train)
model_c.score(X_test,y_test)
```

```python
model_c=SVC(C=10)
model_c.fit(X_train,y_train)
model_c.score(X_test,y_test)
```

```python
model_g=SVC(gamma=10)
model_g.fit(X_train,y_train)
model_g.score(X_test,y_test)
```

```python
model_kernel=SVC(kernel='linear')
model_kernel.fit(X_train,y_train)
model_kernel.score(X_test,y_test)
```
